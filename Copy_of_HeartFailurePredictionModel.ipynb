 **Since our dataset is imbalanced, (more inclincation towards 1) it is necessary to balance it. In a way that 1 and 0 are both equated. As for the same, we employ SMOTE analysis to ensure the dataset which we are using consists of equal 1's and 0's. (This is also called oversampling )**

# **Train-Test Split**

```python
X = data.loc[:, data.columns != 'DEATH_EVENT']
Y = data['DEATH_EVENT'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.30,random_state=0)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for better aesthetics
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(8, 6))
ax = sns.countplot(x=Y_train_smote, palette='viridis', edgecolor='black')

plt.title('Class Distribution After SMOTE', fontsize=16, weight='bold')

plt.xlabel('Final Output', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.xticks(ticks=[0, 1], labels=['0.0', '1.0'], fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.xticks(rotation=0)

plt.tight_layout()  # To ensure no clipping of elements
plt.show()
```
